```python
# tell which eeg state this person is in
fs = 160.6
yf = np.absolute(scipy.fftpack.fft(arr2))
xf = np.absolute(np.fft.fftfreq(len(arr2), 1/fs))
yf1 = np.absolute(scipy.fftpack.fft(arr1))
xf1 = np.absolute(np.fft.fftfreq(len(arr1),1/fs))

#define eeg band
eeg_bands = {'Delta': (0, 4),
             'Theta': (4, 8),
             'Alpha': (8, 12),
             'Beta': (12, 30),
             'Gamma': (30, 45)}   

#Take the mean of the fft amplitude for each EEG band
eeg_band_fft = dict()
for band in eeg_bands:  
    freq_ix = np.where((xf >= eeg_bands[band][0]) & 
                       (xf <= eeg_bands[band][1]))[0]
    eeg_band_fft[band] = np.max(yf[freq_ix])
    df = pd.DataFrame(columns=['band', 'val'])
df['band'] = eeg_bands.keys()
df['val'] = [eeg_band_fft[band] for band in eeg_bands]
fig, ax = plt.subplots(1,2, figsize = [16,14])
df.plot(ax = ax[0],kind='bar',x='band', y='val',figsize=(10,5),legend = False,fontsize = 15,color='g')
ax[0].set_xlabel("EEG band Breath",fontsize = 15)
ax[0].set_ylabel("Max band Amplitude",fontsize = 15)


#Take the mean of the fft amplitude for each EEG band
eeg_band_fft = dict()
for band in eeg_bands:  
    freq_ix = np.where((xf1 >= eeg_bands[band][0]) & 
                       (xf1 <= eeg_bands[band][1]))[0]
    eeg_band_fft[band] = np.max(yf1[freq_ix])
    df2 = pd.DataFrame(columns=['band', 'val'])
df['band'] = eeg_bands.keys()
df['val'] = [eeg_band_fft[band] for band in eeg_bands]
df.plot(ax = ax[1],kind ='bar',x='band', y='val',figsize=(18,7),legend = False,fontsize = 15,color='b')
ax[1].set_xlabel("EEG band Body scan",fontsize = 15)
ax[1].set_ylabel("Max band Amplitude",fontsize = 15)
plt.suptitle('Meditation EEG Band for full data',fontsize = '20')
fig.tight_layout(pad = 4)
```
